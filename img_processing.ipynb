```python
import cv2 as cv
from google.colab.patches import cv2_imshow as cv_imshow
import skimage.data as data
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = data.rocket()
cv_imshow(img)
print(img.shape)
```

```python
print(img[20,50])#printing rgb vale of certain pixel loaction
#B , G ,R value respectively
print(img[20,50,0])
print(img[20,50,1])
print(img[20,50,2])
```

```python
img2=img+100
cv_imshow(img+100)

#note images must e of same size
cv_imshow(img+img2)#addition

cv_imshow(img-img2)#subtraction

cv_imshow(img*img2)#multiplying images 

cv_imshow(img*0.5)#multiplying with a scaling factor

cv_imshow(img/img2)
```

```python
img2=cv.resize(img,(256,256))
img2=cv.cvtColor(img2,cv.COLOR_BGR2GRAY)
cv_imshow(img2)

plt.hist(img2.ravel(),256,[0,255])
```

```python
"""for i in range(256):
  for j in range(255):
    if img2[i,j] < 150:
      img2[i,j] =img2[i,j]/4 + 1
"""
img3 = cv.blur(img2,(10,10))
cv_imshow(img3)
dgr_val= img3/img2
plt.hist(dgr_val.ravel(),256)
cv_imshow(img3/dgr_val)
```

```python
img_gaus_blurr= cv.GaussianBlur(img2,(3,3),8)
cv_imshow(img_gaus_blurr)
cv_imshow(img_gaus_blurr/(img_gaus_blurr/img2))
```

```python
img_median_blur =cv.medianBlur(img2, 7)
cv_imshow(img_median_blur)
cv_imshow(img_median_blur/(img_median_blur/img2))
```

```python
from skimage.util import random_noise


s_p = random_noise(img2,mode='s&p')
s_p = np.array(255*s_p, dtype = 'uint8')
cv_imshow(s_p)
cv_imshow(cv.fastNlMeansDenoising(s_p,None,10,7,21))
```

```python
gaus_noise = random_noise(img2,var=0.025**2)
gaus_noise = np.array(255*gaus_noise, dtype = 'uint8')
cv_imshow(gaus_noise)
cv_imshow(cv.fastNlMeansDenoising(gaus_noise,None,8,3,10))
```
